```python
import gym
import time
import random
import numpy as np
from gym.envs.toy_text.blackjack import BlackjackEnv
#创建环境
env = BlackjackEnv()

#MC的On-Policy的采样
def Sample(env,Q,epsilon,Policy=False):
    Obs,All_Reward,All_Actions=[],[],[]
    Obtion=env.reset()
    Obs.append(Obtion)
    Over=False
    while not Over:
        if Policy:
            Select_Action=random.randint(0,1)
        else:
            if random.random()>=epsilon:
                Select_Action=Greedy(Q,Obtion)
            else:
                Select_Action=random.randint(0, 1)
        Obtion,TheR,Over,_=env.step(Select_Action)
        if not Over:
            Obs.append(Obtion)
        All_Reward.append(TheR)
        All_Actions.append(Select_Action)
    return {'obs':Obs,'actions':All_Actions,'rewards':All_Reward}

#epsilon-greedy选择概率
def Greedy(Q,Obs):
    Gid=(Obs[0],Obs[1],int(Obs[2]))
    return np.argmax(Q[Gid])

#计算Q表
def Compute_Q(env,epsilon,C_Num=5000):
    Obs_space=env.observation_space
    Action_space=env.action_space
    k=[some.n for some in Obs_space]+[Action_space.n]
    Q=np.zeros(k)
    Num_Q=np.zeros(k)
    for i in range(C_Num):
        data=Sample(env,Q,epsilon)
        Q,Num_Q=MC(data,Q,Num_Q)
    return Q

#蒙特卡罗算法
def MC(D,Q,Num_Q):
    T,Gama=0,1
    TheRewards,TheActions,Obs=D['rewards'],D['actions'],D['obs']
    for i in range(len(Obs)-1,-1,-1): 
        T=T*Gama+TheRewards[i]
        if TheActions[i] in TheActions[:i] and Obs[i] in Obs[:i]: 
            continue
        id=(Obs[i][0],Obs[i][1],int(Obs[i][2]),TheActions[i])
        Num_Q[id]=Num_Q[id]+1
        Q[id]=Q[id]+(T-Q[id])/Num_Q[id]
    return Q,Num_Q

#检测最终结果
def main(Q,epsilon,Policy=False,Num=5000):
    Win=0
    Draw=0
    Lose=0
    for i in range(Num):
        data = Sample(env,Q,epsilon,Policy)
        if data['rewards'][-1]<0:
            Lose=Lose+1
        elif data['rewards'][-1]>0:
            Win=Win+1
        else:
            Draw=Draw+1
    print('当epsilon为: '+str(epsilon))
    print('胜率为: '+str(Win/Num)[:5]+' 负率为: '+str(Lose/Num)[:5])
    print('        ')

#不同的epsilon，线性衰减
epsilon=0
for i in range(20):
    epsilon=1-i*0.05
    Q=Compute_Q(env,epsilon,C_Num=5000)
    main(Q,epsilon)





```
