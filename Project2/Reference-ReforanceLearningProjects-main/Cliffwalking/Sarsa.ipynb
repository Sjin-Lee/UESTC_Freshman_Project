```python
import time
import gym
import numpy as np
import numpy as py

#创建悬崖环境
env=gym.make("CliffWalking-v0")

#定义一个Sarsa类，便于修改参数
class Sarsa(object):
    def __init__(self,o_dimension,a_dimension,LearningRate,Gama,Epsilon_greed):
        self.lr=LearningRate
        self.Epsilon=Epsilon_greed
        self.Gama=Gama
        self.Q=np.zeros((o_dimension,a_dimension))
        self.dimension=a_dimension

    def Select_Action(self,observation):
        if 1-self.Epsilon<=np.random.uniform(0,1):
            Action=np.random.choice(self.dimension)
        else:
            Action=self.Conjecture_Action(observation)
        return Action

    def Conjecture_Action(self,observation):
        Q_Value=self.Q[observation,:]
        Max_Q=np.max(Q_Value)
        Action=np.random.choice(np.where(Q_Value==Max_Q)[0])
        return Action

    def Update_Q(self,observation,Action,R,New_observation,New_Action,Over):
        Conjecture_Q=self.Q[observation,Action]
        if Over:
            New_Q=R
        else:
            New_Q=R+self.Gama*self.Q[New_observation,New_Action]
            self.Q[observation,Action]=self.Q[observation,Action]+self.lr*(New_Q-Conjecture_Q)

def Episode(env,agent,render):
    Steps=0
    Total_reward=0
    observation=env.reset()
    Action=agent.Select_Action(observation)
    while 1:
        New_observation,R,Over,_=env.step(Action)
        New_Action=agent.Select_Action(New_observation)
        agent.Update_Q(observation,Action,R,New_observation,New_Action,Over)
        Action=New_Action
        observation=New_observation
        Total_reward=Total_reward+R
        Steps=Steps+1
        if render:
            env.render()
        if Over:
            break
    return Total_reward,Steps

agent=Sarsa(
    o_dimension=env.observation_space.n,
    a_dimension=env.action_space.n,
    LearningRate=0.1,
    Gama=0.9,
    Epsilon_greed=0.1
)

for i in range(1000):
    rewards,steps=Episode(env,agent,False)
    print('第%s次episode，用了%s步，总回报为%.2f'%(i,steps,rewards))
```
