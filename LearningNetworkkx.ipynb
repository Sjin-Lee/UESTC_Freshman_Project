```python
import warnings
warnings.filterwarnings("ignore")

import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.Graph() # Create a new Graph
```

```python
G.add_node(1)
G.add_node(2)
G.add_node(3)

nx.draw(G,with_labels=True)
plt.show()
```

```python
G.add_edge(1,2)
G.add_edge(2,3)
nx.draw(G,with_labels=True)
plt.show()
```

```python
# When there are many edges ,it will be much more difficult
# so we can use the followings
G2=nx.Graph()
G2.add_nodes_from(list(range(1,5)))
G2.add_edges_from([[1,2],[2,3],[3,4],[4,1]])
nx.draw(G2,with_labels=True)
plt.show()
```

```python
# use OPI to create random graphs
randG=nx.erdos_renyi_graph(n=10,p=0.5)
nx.draw(randG)
# n:number of edges 
# p connected possibility
plt.show()
```

```python
# Read from data
def readdata(filepath):
    data=[]
    with open(filepath,'r') as f:
        for line in f.readlines():
            data.append(tuple(line.strip(),split('\t')))
    return data
```
